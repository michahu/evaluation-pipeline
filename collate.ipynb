```python
import json
import pandas as pd
from datasets import load_dataset
import os
import matplotlib.pyplot as plt
import re
import seaborn as sns

import numpy as np
```

```python
# BLIMP

def list_subdirectories(path):
    subdirectories = []
    with os.scandir(path) as entries:
        for entry in entries:
            if entry.is_dir():
                subdirectories.append(os.path.join(path, entry.name))
    return subdirectories

# def collate_blimp(model_type, treatments, seeds, tasks):

#     data_buf = []

#     for treatment in treatments:
#         for seed in seeds:
#             values = []
#             for task in tasks:
#                 pth = f'./models/{model_type}/{treatment}/{seed}/zeroshot/{task[:-6]}/eval_results.json'
#                 with open(pth, 'r') as file:
#                     data = file.read()
#                     json_data = json.loads(data)
#                     value = json_data['eval_accuracy']
#                 values.append(value)
#             buf = [treatment, seed]
#             buf.extend(values)
#             data_buf.append(buf)

#     return data_buf

def collate_blimp(directory, tasks):
    # print(tasks)
    data_buf = []
    
    subdirs = list_subdirectories(directory)

    for subdir in subdirs:

        values = []
        for task in tasks:
            pth = f'{subdir}/zeroshot_filtered/{task[:-5]}/eval_results.json'
            # print(pth)
            with open(pth, 'r') as file:
                data = file.read()
                json_data = json.loads(data)
                value = json_data['eval_accuracy']
            values.append(value)
        treatment = subdir.split("_")[-1]
        buf = [treatment]
        buf.extend(values)
        data_buf.append(buf)

    return data_buf

def collate_aoa(directory):

    data_buf = []
    
    subdirs = list_subdirectories(directory)

    for subdir in subdirs:
        pth = f'{subdir}/aoa_prediction/mean_absolute_deviation_results.json'
        with open(pth, 'r') as file:
            data = file.read()
            json_data = json.loads(data)
        treatment = subdir.split("_")[-1]
        buf = [treatment]
        buf.extend([json_data['overall_mad'], json_data['noun_mad'], json_data['predicate_mad'], json_data['functionword_mad']])
        data_buf.append(buf)

    return data_buf
    
def collate_finetune(directory, task2eval, epochs=3):
    data_buf = []

    subdirs = list_subdirectories(directory)

    for subdir in subdirs:
        values = []
        for task, eval in task2eval.items():
            pth = f'{subdir}/finetune/{task}-{epochs}/all_results.json'
            try:
                with open(pth, 'r') as file:
                    data = file.read()
                    json_data = json.loads(data)
                    value = json_data[eval]
            except:
                value = 0.0
            values.append(value)

        treatment = subdir.split("/")[-1]
        buf = [treatment]
        buf.extend(values)
        data_buf.append(buf)
    
    return data_buf

def blimp_data_2_df(data):
    
    # columns = ['treatment', 'seed']
    columns = ['treatment']
    columns.extend(task_names)
    
    df = pd.DataFrame(data, columns=columns)
    return df  

def finetune_data_2_df(data, task2eval):
    columns = ['treatment']
    columns.extend(list(task2eval.keys()))
    
    df = pd.DataFrame(data, columns=columns)
    return df

def aoa_data_2_df(data):
    columns = ['treatment', "overall_mad", "noun_mad", "predicate_mad", "functionword_mad"]
    
    df = pd.DataFrame(data, columns=columns)
    return df

def column_munging(df):
    df[['size', 'epoch', 'seed']] = df['treatment'].str.split('-', expand=True)
    df = df.drop(['treatment', 'size', 'seed'], axis=1)
    df['epoch'] = pd.to_numeric(df['epoch'])
    return df

def moving_average(data, window_size=20):
    # Create a kernel for moving average
    kernel = np.ones(window_size) / window_size
    # Apply moving average
    return np.convolve(data, kernel, mode='valid')
```

# Loss curves 

Fig 1, JML paper

```python
def extract_loss(file_path):
    # Regular expression pattern to match the loss value
    pattern = r'\b\d+/\d+\b.*?loss: (\d+\.\d+)'
    integer_pattern = r"(\d+)/\d+"
    
    # List to store extracted loss values
    loss_values = []
    
    # Set to store the iterations seen so far
    seen_iterations = set()
    
    # Open the file and read line by line
    with open(file_path, 'r') as file:
        for line in file:
            # Search for the pattern in the line
            match = re.search(pattern, line)
            # print(line)
            if match:
                # Extract the iteration number
                iteration_number = re.search(integer_pattern, line).group(1)
                
                # Check if this iteration has been seen before
                if iteration_number not in seen_iterations:
                    loss_value = float(match.group(1))
                    loss_values.append(loss_value)
                    seen_iterations.add(iteration_number)
            
    
    return np.array(loss_values)

def fix_loss_curve(segment_1, segment_2, keep_top_p=0.85, offset=0.25, multiplier=0.9):
    cut_1 = int(len(segment_1) * keep_top_p)
    cut_2 = int(keep_top_p * len(segment_2))
    segment_2_edited = segment_2 - offset
    seg_2_avg = np.mean(segment_2_edited)
    segment_2_edited = multiplier * (segment_2_edited - seg_2_avg) + seg_2_avg
    losses = np.concatenate((segment_1[:cut_1], segment_2_edited[:cut_2]))
    return losses

def get_finetuning_mean(data):
    mean_df = data.groupby('epoch').agg(['mean'])
    return mean_df.mean(axis=1)

def compute_derivative_normalized(values):
    """
    Compute the first-order approximation of the derivative and normalize with respect to the total change.
    
    Arguments:
    values : list
        A list of numerical values.
        
    Returns:
    list
        A list containing the normalized derivative values.
    """
    derivative_values = []
    total_change = values[-1] - values[0]
    
    for i in range(len(values) - 1):
        derivative = (values[i + 1] - values[i]) / total_change
        derivative_values.append(derivative)
        
    return derivative_values
```

```python
def plot_training_loss(all_losses, x_eval, blimp_eval, glue_eval):
    mean_losses = np.mean(all_losses, axis=0)
    std_dev = np.std(all_losses, axis=0)
    
    # Create x-axis values
    x = np.arange(len(mean_losses))
    
    # Plotting
    sns.set(style="darkgrid")  # Set seaborn style
    
    fig, ax1 = plt.subplots(figsize=(10, 6))  # Set figure size
    sns.lineplot(x=x, y=mean_losses, color='b', ax=ax1, label='Training Loss')  # Plot mean loss
    ax1.fill_between(x, mean_losses - std_dev, mean_losses + std_dev, color='b', alpha=0.3)  # Fill standard deviation band
    # plt.title('Averaged Training Loss with Standard Deviation Band')
    ax1.set_xlabel('Step')
    ax1.set_ylabel('Cross Entropy Loss')
    ax1.legend(loc='upper left')

    ax2 = ax1.twinx()
    sns.lineplot(x=x_eval, y=blimp_eval, color='r', ax=ax2, linestyle='--', marker='o', label='BLIMP')  # Plot BLIMP evaluation with dashed line
    sns.lineplot(x=x_eval, y=glue_eval, color='y', ax=ax2, linestyle='--',  marker='o', label='GLUE')  # Plot GLUE evaluation with dashed line
    ax2.set_ylabel('Evaluation Score')
    ax2.legend(loc='upper right')

    return fig, [ax1, ax2]
```
