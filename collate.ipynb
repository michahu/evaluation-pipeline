```python
# BLIMP

def collate_blimp(model_type, treatments, seeds, tasks):

    data_buf = []

    for treatment in treatments:
        for seed in seeds:
            values = []
            for task in tasks:
                pth = f'./models/{model_type}/{treatment}/{seed}/zeroshot/{task[:-6]}/eval_results.json'
                with open(pth, 'r') as file:
                    data = file.read()
                    json_data = json.loads(data)
                    value = json_data['eval_accuracy']
                values.append(value)
            buf = [treatment, seed]
            buf.extend(values)
            data_buf.append(buf)

    return data_buf

def collate_glue(model_type, treatments, seeds, task2eval):
    data_buf = []

    for treatment in treatments:
        for seed in seeds:
            values = []
            for task, eval in task2eval.items():
                pth = f'./models/{model_type}/{treatment}/{seed}/finetune/{task}-3/all_results.json'
                with open(pth, 'r') as file:
                    data = file.read()
                    json_data = json.loads(data)
                    value = json_data[eval]
                values.append(value)
            buf = [treatment, seed]
            buf.extend(values)
            data_buf.append(buf)
    
    return data_buf

def blimp_data_2_df(data):
    
    columns = ['treatment', 'seed']
    columns.extend(task_names)
    
    df = pd.DataFrame(data, columns=columns)
    return df    
```
