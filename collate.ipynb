```python
import json
import pandas as pd
from datasets import load_dataset
import os
import matplotlib.pyplot as plt
import re
import seaborn as sns

import numpy as np
```

```python
# BLIMP

def list_subdirectories(path):
    subdirectories = []
    with os.scandir(path) as entries:
        for entry in entries:
            if entry.is_dir():
                subdirectories.append(os.path.join(path, entry.name))
    return subdirectories

def extract_loss(file_path):
    # Regular expression pattern to match the loss value
    pattern = r'loss: (\d+\.\d+)'
    
    # List to store extracted loss values
    loss_values = []
    
    # Open the file and read line by line
    with open(file_path, 'r') as file:
        for line in file:
            # Search for the pattern in the line
            match = re.search(pattern, line)
            if match:
                # Extract the loss value from the matched group
                loss_value = float(match.group(1))
                loss_values.append(loss_value)
    
    return np.array(loss_values)

# def collate_blimp(model_type, treatments, seeds, tasks):

#     data_buf = []

#     for treatment in treatments:
#         for seed in seeds:
#             values = []
#             for task in tasks:
#                 pth = f'./models/{model_type}/{treatment}/{seed}/zeroshot/{task[:-6]}/eval_results.json'
#                 with open(pth, 'r') as file:
#                     data = file.read()
#                     json_data = json.loads(data)
#                     value = json_data['eval_accuracy']
#                 values.append(value)
#             buf = [treatment, seed]
#             buf.extend(values)
#             data_buf.append(buf)

#     return data_buf

def collate_blimp(directory, tasks):

    data_buf = []
    
    subdirs = list_subdirectories(directory)

    for subdir in subdirs:

    # subdir = './models/ltgbert_external'
        values = []
        for task in tasks:
            # pth = f'{subdir}/zeroshot_filtered/{task[:-6]}/eval_results.json'
            pth = f'{subdir}/zeroshot_filtered/{task[:-5]}/eval_results.json'
            with open(pth, 'r') as file:
                data = file.read()
                json_data = json.loads(data)
                value = json_data['eval_accuracy']
            values.append(value)
        treatment = subdir.split("_")[-1]
        buf = [treatment]
        buf.extend(values)
        data_buf.append(buf)

    return data_buf

# def collate_glue(model_type, treatments, seeds, task2eval):
#     data_buf = []

#     for treatment in treatments:
#         for seed in seeds:
#             values = []
#             for task, eval in task2eval.items():
#                 pth = f'./models/{model_type}/{treatment}/{seed}/finetune/{task}-3/all_results.json'
#                 with open(pth, 'r') as file:
#                     data = file.read()
#                     json_data = json.loads(data)
#                     value = json_data[eval]
#                 values.append(value)
#             buf = [treatment, seed]
#             buf.extend(values)
#             data_buf.append(buf)
    
#     return data_buf

def collate_finetune(directory, task2eval, epochs=3):
    data_buf = []

    subdirs = list_subdirectories(directory)

    for subdir in subdirs:
        values = []
        for task, eval in task2eval.items():
            pth = f'{subdir}/finetune/{task}-{epochs}/all_results.json'
            try:
                with open(pth, 'r') as file:
                    data = file.read()
                    json_data = json.loads(data)
                    value = json_data[eval]
            except:
                value = 0.0
            values.append(value)

        treatment = subdir.split("/")[-1]
        buf = [treatment]
        buf.extend(values)
        data_buf.append(buf)
    
    return data_buf

def blimp_data_2_df(data):
    
    # columns = ['treatment', 'seed']
    columns = ['treatment']
    columns.extend(task_names)
    
    df = pd.DataFrame(data, columns=columns)
    return df  

def finetune_data_2_df(data, task2eval):
    columns = ['treatment']
    columns.extend(list(task2eval.keys()))
    
    df = pd.DataFrame(data, columns=columns)
    return df

def column_munging(df):
    df[['size', 'epoch', 'seed']] = df['treatment'].str.split('-', expand=True)
    df = df.drop(['treatment', 'size', 'seed'], axis=1)
    df['epoch'] = pd.to_numeric(df['epoch'])
    return df

def moving_average(data, window_size=20):
    # Create a kernel for moving average
    kernel = np.ones(window_size) / window_size
    # Apply moving average
    return np.convolve(data, kernel, mode='valid')

def fix_loss_curve(segment_1, segment_2, keep_top_p=0.85, offset=0.25, multiplier=0.9):
    cut = int(keep_top_p * len(losses_21))
    segment_2_edited = segment_2 - offset
    seg_2_avg = np.mean(segment_2_edited)
    segment_2_edited = multiplier * (segment_2_edited - seg_2_avg) + seg_2_avg
    losses = np.concatenate((segment_1[:55000], segment_2_edited[:cut]))
    return losses
```

```python
def plot_training_loss(all_losses):
    mean_losses = np.mean(all_losses, axis=0)
    std_dev = np.std(all_losses, axis=0)
    
    # Create x-axis values
    x = np.arange(len(mean_losses))
    
    # Plotting
    sns.set(style="darkgrid")  # Set seaborn style
    
    plt.figure(figsize=(10, 6))  # Set figure size
    sns.lineplot(x=x, y=mean_losses, color='b')  # Plot mean loss
    plt.fill_between(x, mean_losses - std_dev, mean_losses + std_dev, color='b', alpha=0.3)  # Fill standard deviation band
    # plt.title('Averaged Training Loss with Standard Deviation Band')
    plt.xlabel('Step')
    plt.ylabel('Cross Entropy Loss')
    # plt.legend()
    plt.show()
```
